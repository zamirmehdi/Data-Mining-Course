<h1>قوانین انجمنی</h1>

قوانین انجمنی روابط و وابستگی‌های متقابل بین مجموعه بزرگی از اقلام داده‌ای را نشان می‌دهند. 

مثال متداول در رابطه با کشف قوانین انجمنی "تحلیل سبد خرید" است.  در این فرایند با توجه به اقلام مختلفی که مشتریان در سبد خریدشان قرار می‌دهند، عادات و رفتار خرید مشتریان مورد تحلیل قرار می‌گیرد و می‌توان با شناسایی ارتباط بین محصولات، الگوهای تکرار شونده در حین خرید را بدست آورد.

سه پارامتر مهم:
- Support  میزان محبوبیت یک مجموعه آیتم را با توجه به دفعات حضور در تراکنش‌ها نشان می‌دهد.
- Confidence  احتمال خرید آیتم y در صورت خرید آیتم x را نشان می‌دهد. x -> y
- Lift   ترکیب دو پارامتر بالا می‌باشد.

برای پیاده سازی قوانین انجمنی در این تمرین، از الگوریتم Apriori که یکی از محبوب‌ترین و کارآمدترین الگوریتم‌های این حوزه است استفاده می‌کنیم.


<font color='Green'> پرسش:  تاثیر مقادیر مختلف پارامتر Lift را در احتمال وقوع نتیجه بررسی کنید.  </font>


</div>

پاسخ: 
لیفت یک معیار برای اندازه‌گیری همبستگی است که به شرح زیر ارائه می شود. وقوع مجموعه اقلام A مستقل از وقوع مجموعه اقلام B است
اگر

 P(A ∪ B) = P(A)P(B); در غیر این صورت، مجموعه آیتم های A و B به عنوان رویداد وابسته و همبسته هستند. این تعریف را می توان به راحتی به بیش از دو مجموعه آیتم تعمیم داد. میزان lift بین وقوع A و B را می توان با محاسبه اندازه گیری کرد:
lift(A,B)= P(A∪B) / P(A)P(B)

اگر مقدار lift کمتر از 1 باشد؛ وقوع A با وقوع B همبستگی منفی دارد، به این معنی که وقوع یکی به احتمال زیاد منجر به عدم وجود دیگری می شود. اگر مقدار lift بزرگتر از 1 باشد، A و B همبستگی مثبت دارند، به این معنی که وقوع یکی دلالت بر وقوع دیگری دارد. و اگر مقدار حاصل برابر با 1 باشد، A و B مستقل هستند و هیچ ارتباطی بین آنها وجود ندارد.

پس مقدار بالای lift  ارتباط مثبت زیاد دو آیتم را نشان می‌دهد و مقدار کم lift  ارتباط ضعیف آن دو را!

<h1>الگوریتم Apriori</h1>

روش کار الگوریتم به این صورت است که یک  حداقل مقدار پشتیبانی در نظر گرفته    می‌شود و تکرارها با مجموعه آیتم‌های مکرر اتفاق می‌افتند. در صورتیکه  مجموعه‌ها و زیرمجموعه‌ها مقدار پشتیبانی کمتر از حد آستانه داشته باشند حذف می‌شوند. این روند تا زمانیکه امکان حذف وجود نداشته باشد ادامه می‌یابد. 

در این بخش از تمرین می‌خواهیم الگوریتم apriori را بر روی مجموعه داده Hypermarket_dataset که شامل سفارشات خرید افراد از فروشگاه های مواد غذایی است اعمال کنیم.

<h1>آماده‌سازی داده</h1>

<font color='Green'>پرسش: برای شروع کار باید داده‌های دیتاست را در قالب ماتریس پراکنده که محصولات  خریداری شده در ستون‌ و شماره سفارشات خرید به عنوان شاخص (index) هستند آماده کنید. 


برای راحتی کار محصولات خریداری شده در هر سفارش را با اعداد 0,1 کدگذاری کنید.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# your code here
# ...

dataset = pd.read_csv("Hypermarket_dataset.csv", header = None)
dataset.iloc[1: , :]

columns = []
j = 0
for i in dataset.iloc[:, 2]:
  if j == 0:
    j+=1
    continue
  temp_columns = i
  columns += [temp_columns]
columns = list(set(columns)) # remove redundant columns

transaction_dict = {}
i = 0
for index, row in dataset.iterrows():
  if i == 0:
    i+=1
    continue
  index = row[0]
  items = row[2]
  if index in transaction_dict:
    transaction_dict[index] += [items]
  else:
    transaction_dict[index] = [] 
    transaction_dict[index] += [items]

matrix = []
for key, value in transaction_dict.items():
  new_item_list = [0]*len(columns)
  for temp_val in value:
    new_item_list[columns.index(temp_val)] = 1
  matrix.append(new_item_list)

# print(transaction_dict['1000'])

indexes = list(transaction_dict.keys())
output_dataframe = pd.DataFrame(matrix, index=indexes, columns=columns)
output_dataframe.sort_index()



```


<div dir=rtl >
<h1>شناسایی الگوهای پر تکرار</h1>

<font color='Green'> پرسش:  با اعمال الگوریتم apriori و به ازای مقدار min_support = 0.07 تمامی الگوهای پر تکرار را تولید نمایید. 

</div>




```python
# your code here
# ...
frequent_itemsets = apriori(output_dataframe, min_support=0.07, use_colnames=True)
frequent_itemsets['k-freq'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
# frequent_itemsets = frequent_itemsets.sort_values('support')
frequent_itemsets
```


<div dir=rtl >
<h1>استخراج قوانین انجمنی</h1>


<font color='Green'> پرسش: تابعی بنویسید که با گرفتن دو ورودی confidence و lift قوانین انجمنی حاصل را در خروجی نشان دهد. </br>
برای دو حالت خروجی خود را در گزارش ثبت کنید.


</div>




```python
# your code here
# ...

def extract_rules(confidence_or_lift, amount):
  rules = association_rules(frequent_itemsets, metric=confidence_or_lift, min_threshold=amount)
  if confidence_or_lift == 'lift':
    rules = rules.sort_values("lift",ascending=False)
  else:
    rules = rules.sort_values("confidence",ascending=False)

  return rules
```

```python
extract_rules('confidence', 0.07)
```

```python
extract_rules('lift', 1.2)
```
